Extract All the URLs from base URL.

```python
from rfc3986 import urlparse


def extract_urls_from_url(start_url):
        visited_urls = set()
        queue = [start_url]
        queued_urls = set([start_url])
        while queue:
            current_url = queue.pop(0)
            if current_url not in visited_urls:
                try:
                    response = requests.get(current_url)
                    if response.status_code == 200:
                        visited_urls.add(current_url)
                        soup = BeautifulSoup(response.content, "html.parser")
                        links = soup.find_all("a", href=True)
                        for link in links:
                            absolute_url = urljoin(current_url, link["href"])
                            parsed_url = urlparse(absolute_url)
                            if parsed_url.scheme in ['http', 'https'] and \
                               'mailto:' not in absolute_url and \
                               'tel:' not in absolute_url and \
                               absolute_url.startswith(start_url):
                                if absolute_url not in visited_urls and \
                                   absolute_url not in queued_urls:
                                    queue.append(absolute_url)
                                    queued_urls.add(absolute_url)
                except Exception as e:
                    print(f"An error occurred: {e}")
        return visited_urls
```

Convert all the URLs to Image

```python
import os
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from urllib.parse import urlparse
import time
options = Options()
options.add_argument('--headless')  # Set the browser to run in headless mode
# Initialize the Chrome driver
driver = webdriver.Chrome(options=options)
def capture_screenshot(url):
    try:
        driver.get(url)
        print(f"Capturing screenshot for: {url}")

        # Parse the URL to extract domain and path
        parsed_url = urlparse(url)
        domain = parsed_url.netloc
        path = parsed_url.path

        # Format the domain and path for the folder name
        domain = domain.replace('.', '-')  # Replacing dots with hyphens for the folder name
        path = path.replace('/', '-')  # Replacing slashes with hyphens for the folder name
        if not os.path.exists("WebImages"):
            os.makedirs("WebImages")
        # Create the folder to save screenshots for this URL
        folder_name = f"WebImages/{domain}{path}"
        if not os.path.exists(folder_name):
            os.makedirs(folder_name)

        # Set initial window size for capturing screenshots
        window_width = 1920
        window_height = 1080
        driver.set_window_size(window_width, window_height)

        # Get total page height to scroll through
        total_height = driver.execute_script("return document.body.scrollHeight")

        # Take individual screenshots and save them in the respective folder
        screenshot_counter = 1
        scroll_offset = 0
        while scroll_offset < total_height:
            driver.execute_script(f"window.scrollTo(0, {scroll_offset});")
            screenshot_name = f"{folder_name}/{screenshot_counter}.png"
            driver.save_screenshot(screenshot_name)
            screenshot_counter += 1

            # Calculate the height of the visible part of the document
            visible_height = driver.execute_script("return window.innerHeight")
            scroll_offset += visible_height  # Move the offset by the visible height

            # Give time for the page to load if needed
            time.sleep(1)

        print(f"Individual screenshots saved for: {url} in folder: {folder_name}")
    except Exception as e:
        print(url,e)
```

Use tesseract-ocr for processing image

```python
client = boto3.client('textract',region_name='****',aws_access_key_id='**************',aws_secret_access_key='*******************')
def ocr(image):
    # Convert the cv2 image to bytearray (using JPEG encoding here)
    retval, buffer = cv2.imencode('.jpg', image)

    # Check if the image was successfully encoded
    if not retval:
        print("Failed to encode the image.")
    else:
        # Convert the buffer to a bytearray
        img = bytearray(buffer)

    response = client.detect_document_text(Document={"Bytes": img})
    ocr_result = ""
    for item in response["Blocks"]:
        if item["BlockType"] == "LINE":
            ocr_result += " " + item["Text"]
    return ocr_result
```

Extract text from image using AWS Tesseract and Upsert it into Database

```python

main_folder_path = "WebImages"
docs = []
for folder in os.listdir(main_folder_path):
    subfolder_path = os.path.join(main_folder_path, folder)
    data = []
    if os.path.isdir(subfolder_path):
        images = sorted([file for file in os.listdir(subfolder_path) if file.endswith('.png')])
        print(images)
        for image_name in tqdm(images, desc=f"Processing {folder}"):
            image_path = os.path.join(subfolder_path, image_name)

            image = cv2.imread(image_path)
            try:
                results = pytesseract.image_to_osd(image, output_type=pytesseract.Output.DICT)
                image = imutils.rotate_bound(image, angle=results["rotate"])

                ocr_result = ocr(image)  # Your OCR function
                data.append(ocr_result) 
            except pytesseract.TesseractError:
                continue
    ocr_result = " ".join(data)
    docs.append(Document(page_content=ocr_result, metadata={"source": folder}))

text_splitter = RecursiveCharacterTextSplitter(chunk_size=2500,
                                                chunk_overlap=200
                                                  )
print("results: ",docs)
texts = text_splitter.split_documents(docs)
print("Splitted document: ",texts)
docsearch = Pinecone.from_existing_index(index_name, embeddings)
docsearch.add_documents(texts)


#To delete the Directory
#shutil.rmtree("WebImages")

```
